#### 1.Checking the Sales on Region wise


```python
# plotting a bar chart for Sales vs Region
#Grouping the data by Region
df_sales_Region=df_sd.groupby(['Region']).sum()[['Sales']]
df_sales_Region.plot(kind="bar",figsize=(5,5))
plt.xlabel('Region',fontsize=20,color='green') # adding label on thr x-axis
plt.ylabel('Sales',fontsize=20,color='green')
plt.title("Sales vs Region",fontsize=25,color="red") #adding the title to the plot
plt.xticks(rotation=360)
plt.show()
df_sales_Region
```

##### From above graph we can see that west region has the most sales and south region having smaller sales

### 2. check sales vs  state

```python
df_state_sales= df_sd.groupby(['State'], as_index=False).sum()
df_state_sales.sort_values(by='Sales', ascending=False, inplace=True)


plt.figure(figsize=(25,15))
plt.bar(df_state_sales['State'], df_state_sales['Sales'])
plt.xlabel("State")
plt.ylabel("Sales")
plt.title("Sales Generated by State")
plt.xticks(rotation=90)

plt.show()
df_state_sales
```

##### from above graph we can observe that the top 5 States having the highest sales
##### 1.california 2.New York 3.Texas 4.Washington 5.Pennsylvania

### 3.check the profit vs region and profit vs state

```python
df_profit_Region=df_sd.groupby(['Region']).sum()['Profit']

df_profit_Region.plot(kind="bar",figsize=(10,5))

plt.title("Profit vs Region",fontsize=25,color='green')
plt.xlabel('Region',fontsize=20,color='black')
plt.ylabel('profit',fontsize=20,color='black')
plt.xticks(rotation=360)
plt.show()
df_profit_Region
```

```python
df_state_profit= df_sd.groupby(['State'], as_index=False).sum()
df_state_profit.sort_values(by='Profit', ascending=False, inplace=True)


plt.figure(figsize=(25,15))
plt.bar(df_state_profit['State'], df_state_profit['Profit'])
plt.xlabel("State")
plt.ylabel("profit")
plt.title("Sales Generated by State")
plt.xticks(rotation=90)

plt.show()
df_state_profit
```

##### from above two graph we can observe that west region has the highest profit and central region having the lowest profit
##### and we can observe that the top 3 states that has highest profit are 1.california 2.New york 3.washington but texas and pennsylvania has highest sales why both has the negative profit

```python
df_ship_mode_sales=df_sd.groupby(['Ship Mode']).sum()['Sales']
df_ship_mode=['Second Class', 'Standard Class', 'First Class', 'Same Day']
plt.pie(df_ship_mode_sales,labels=df_ship_mode,autopct='%1.f%%')
plt.title("ship Mode vs sales")
plt.show()
```

```python
df_ship_mode_profit=df_sd.groupby(['Ship Mode']).sum()['Profit']
df_ship_mode=['Second Class', 'Standard Class', 'First Class', 'Same Day']
plt.pie(df_ship_mode_profit,labels=df_ship_mode,autopct='%1.f%%')
plt.title("ship Mode vs Profit")
plt.show()
```

```python
df_segment_sales=df_sd.groupby(['Segment']).sum()['Sales']
df_segment=['Consumer', 'Corporate', 'Home Office']
plt.pie(df_segment_sales,labels=df_segment,autopct='%1.f%%')
plt.title("segment vs sales")
plt.show()
df_segment_sales
```

```python
df_segment_profit=df_sd.groupby(['Segment']).sum()['Profit']
df_segment=['Consumer', 'Corporate', 'Home Office']
plt.pie(df_segment_profit,labels=df_segment,autopct='%1.f%%')
plt.title("segment vs profit")
plt.show()
```
